```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
sns.set_theme(style="darkgrid")
print("✅ Libraries loaded successfully")
```

```python
data_path = "../data/raw/"

# Load all CSV files in the folder
all_files = [f for f in os.listdir(data_path) if f.endswith('.csv')]
print(f"Found {len(all_files)} files: {all_files}")

df_list = []
for file in all_files:
    temp = pd.read_csv(os.path.join(data_path, file), encoding='utf-8', low_memory=False)
    temp['source_file'] = file  # track which file each row came from
    df_list.append(temp)

df = pd.concat(df_list, ignore_index=True)
print(f"\n✅ Combined dataset shape: {df.shape}")
print(f"Rows: {df.shape[0]:,} | Columns: {df.shape[1]}")
```

```python
print("=== First 5 rows ===")
df.head()
```

```python
print("=== Column names ===")
print(df.columns.tolist())
```

```python
print("=== Data types ===")
print(df.dtypes)
```

```python
df.columns = df.columns.str.strip()
print("✅ Column names cleaned")
print(df.columns.tolist())
```

```python
# The label column is typically 'Label' in this dataset
target_col = 'Label'

print(f"=== Unique attack types in '{target_col}' ===")
print(df[target_col].value_counts())
```

```python
plt.figure(figsize=(14, 6))
attack_counts = df[target_col].value_counts()
colors = sns.color_palette("husl", len(attack_counts))

bars = plt.bar(attack_counts.index, attack_counts.values, color=colors, edgecolor='black', linewidth=0.5)

plt.title("Attack Type Distribution in Dataset", fontsize=16, fontweight='bold')
plt.xlabel("Attack Type", fontsize=12)
plt.ylabel("Number of Records", fontsize=12)
plt.xticks(rotation=45, ha='right')

# Add count labels on top of bars
for bar, count in zip(bars, attack_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
             f'{count:,}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.savefig("../reports/attack_distribution.png", dpi=150)
plt.show()
print("✅ Chart saved to reports/")
```

```python
print("=== Class distribution (%) ===")
dist = df[target_col].value_counts(normalize=True) * 100
print(dist.round(2))

benign_pct = dist.get('BENIGN', 0)
print(f"\n⚠️  BENIGN traffic = {benign_pct:.1f}% of all data")
print("→ This is a highly imbalanced dataset — we'll handle this in Step 3")
```

```python
print("=== Missing Values ===")
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'missing_count': missing, 'missing_%': missing_pct})
missing_df = missing_df[missing_df['missing_count'] > 0].sort_values('missing_%', ascending=False)

if len(missing_df) == 0:
    print("✅ No missing values found!")
else:
    print(missing_df)
    print(f"\nTotal columns with missing values: {len(missing_df)}")
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

inf_count = np.isinf(df[numeric_cols]).sum().sum()
print(f"Infinite values found: {inf_count}")

if inf_count > 0:
    inf_cols = np.isinf(df[numeric_cols]).sum()
    print("\nColumns with infinite values:")
    print(inf_cols[inf_cols > 0])
```

```python
dupes = df.duplicated().sum()
print(f"Duplicate rows: {dupes:,} ({(dupes/len(df)*100):.2f}%)")
```

```python
print("=== Statistical Summary ===")
df.select_dtypes(include=[np.number]).describe().T.round(2)
```

```python
# Select top 20 numeric features for readability
numeric_df = df.select_dtypes(include=[np.number])
top_features = numeric_df.columns[:20]

plt.figure(figsize=(16, 12))
corr_matrix = numeric_df[top_features].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, mask=mask, annot=False, cmap='coolwarm',
            linewidths=0.5, fmt='.2f', cbar_kws={'shrink': 0.8})

plt.title("Feature Correlation Heatmap (Top 20 Features)", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig("../reports/correlation_heatmap.png", dpi=150)
plt.show()
```

```python
# Create binary label: 0 = BENIGN, 1 = ATTACK
df['binary_label'] = df[target_col].apply(lambda x: 0 if x == 'BENIGN' else 1)

print("=== Binary Label Distribution ===")
print(df['binary_label'].value_counts())
print("\n0 = Benign (normal traffic)")
print("1 = Attack (any attack type)")
```

```python
# Save the cleaned dataframe info
print("=== Dataset Summary ===")
print(f"Total records      : {len(df):,}")
print(f"Total features     : {df.shape[1]}")
print(f"Attack types       : {df[target_col].nunique()}")
print(f"Benign records     : {(df['binary_label']==0).sum():,}")
print(f"Attack records     : {(df['binary_label']==1).sum():,}")
print(f"\nAttack types found : {df[target_col].unique().tolist()}")

# Save a small sample for quick testing later
df.sample(10000, random_state=42).to_csv("../data/processed/sample_10k.csv", index=False)
print("\n✅ Sample of 10,000 rows saved to data/processed/sample_10k.csv")
```
