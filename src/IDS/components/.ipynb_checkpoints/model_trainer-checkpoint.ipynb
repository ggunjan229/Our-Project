```python
import os
import sys
import pickle
from dataclasses import dataclass
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
from src.IDS.exception import CustomException
from src.IDS.logger import logging
from dataclasses import dataclass

```

```python
from src.IDS.utils import read_sql_data

monday_df, tue_fri_df = read_sql_data()

print("Monday shape:", monday_df.shape)
print("Tue-Fri shape:", tue_fri_df.shape)
```

```python

@dataclass
class ModelTrainerConfig:
    iso_model_path: str = os.path.join("artifacts", "isolation.pkl")
    rf_model_path: str = os.path.join("artifacts", "random_forest.pkl")

```

```python
class ModelTrainer:
    def __init__(self):
        self.config = ModelTrainerConfig()

    def train_isolation(self, X_monday):
        try:
            model = IsolationForest(
                n_estimators=300,
                contamination=0.01,
                random_state=42
            )

            model.fit(X_monday)

            with open(self.config.iso_model_path, "wb") as f:
                pickle.dump(model, f)

            logging.info("Isolation Forest trained successfully")

        except Exception as e:
            raise CustomException(e, sys)

    def train_random_forest(self, X, y):
        try:
            X_train, X_test, y_train, y_test = train_test_split(
                X, y,
                test_size=0.2,
                stratify=y,
                random_state=42
            )

            model = RandomForestClassifier(
                n_estimators=500,
                class_weight="balanced",
                random_state=42
            )

            model.fit(X_train, y_train)

            y_pred = model.predict(X_test)

            print("F1 Score:", f1_score(y_test, y_pred))
            print(classification_report(y_test, y_pred))

            with open(self.config.rf_model_path, "wb") as f:
                pickle.dump(model, f)

            logging.info("Random Forest trained successfully")

        except Exception as e:
            raise CustomException(e, sys)
```
